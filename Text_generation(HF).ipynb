```python
from transformers import pipeline

# Load GPT-2 model
generator = pipeline("text-generation", model="gpt2")

# Take user input
prompt = input("Enter your question: ")

# Generate text
#max_new_tokens- A token is a small chunk of text
#num_return_sequences=1 → The model will generate 1 sequence.
#num_return_sequences=3 → The model will generate 3 different sequences,

result = generator(prompt, max_new_tokens=100, num_return_sequences=1)

# Print result
print(result[0]['generated_text'])
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

prompt = "AI will change the future of"
result = generator(prompt, max_new_tokens=20, num_return_sequences=3)

for i, output in enumerate(result):
    print(f"Sequence {i+1}: {output['generated_text']}")
```
